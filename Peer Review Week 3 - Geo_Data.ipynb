```python
import pandas as pd
```

```python
df_wiki = pd.read_csv('wikipage.csv', index_col = 0)
```

```python
df_wiki.head()
```

```python
df_geo = pd.read_csv('Geospatial_Coordinates.csv')
```

```python
df_geo.head()
```

```python
df_merge = df_wiki.merge(df_geo, left_on = 'Postcode', right_on = 'Postal Code')
```

```python
df_merge.drop(columns = ['Postal Code'], inplace = True)
```

```python
df_merge.head()
```

```python
import folium
```

### Generate a map to see how these locations clustered


```python
Toronto_lat = 43.653963
Toronto_lng = -79.387207


map_toronto = folium.Map(location = [Toronto_lat, Toronto_lng], zoom_start = 10)

for lat, lng, borough, neigh in zip(df_merge['Latitude'], df_merge['Longitude'],
                                   df_merge['Borough'], df_merge['Neighbourhood']):
    
    
    
    folium.CircleMarker(
        [lat, lng],
        radius=3,
    
        color='green',
        fill=True,
        fill_color='#3199cc',
        fill_opacity=0.3,
        parse_html=False).add_to(map_toronto)  
    



map_toronto
```
